## Actual Demo!

```python
#imports
import torch
import torchaudio
from transformers import AutoProcessor, SeamlessM4TModel
#Line below used to be from "seamless_communication.models.inference import Translator", was replaced by line below
#from seamless_communication.inference import Translator

#Initialize a Translator object with a multitask model, vocoder on the GPU (Recommend using at least the T4 GPU avaiable on Colab)
#translator = Translator("seamlessM4T_large", vocoder_name_or_card="vocoder_36langs", device=torch.device("cuda:0"))

```

```python
#Load processor and model (This takes a while, go make some coffee or something lol)
processor = AutoProcessor.from_pretrained("facebook/hf-seamless-m4t-large")
model = SeamlessM4TModel.from_pretrained("facebook/hf-seamless-m4t-large")
```

```python
#Implement source and target languages (codes avaiable on the Seamless M4T paper, pages 14 to 15) and input text to be translated
src_lang = "spa" #Spanish
tgt_lang = "jpn" #Japanese
input_text = "Hola, soy Juan, Me gusta estudiar Japonés, vivo en México"
```

```python
# Process the input text
text_inputs = processor(text=input_text, src_lang=src_lang, return_tensors="pt")
```

```python
from IPython.display import Audio
import scipy
```

```python
# Generate the translated text and audio
audio_array_from_text = model.generate(**text_inputs, tgt_lang=tgt_lang)[0].cpu().numpy().squeeze() #Generate Audio from Text
output_tokens = model.generate(**text_inputs, tgt_lang=tgt_lang, generate_speech=False) #Generate text, pay attention to generate_speech=False

#Saving audio
sample_rate = model.config.sampling_rate
scipy.io.wavfile.write("seamless_m4t_out.wav", rate=sample_rate, data=audio_array_from_text) # audio_array_from_audio

translated_text_from_text = processor.decode(output_tokens[0].tolist()[0], skip_special_tokens=True) #Saving and printing translated text
print(f"Translation from text: {translated_text_from_text}")
```

```python
#Display generated audio
Audio(audio_array_from_text, rate=sample_rate)
```
